```python
#tratando os dados
df['ID_CLIENTE_ENCRIPTADO'] = df['ID_CLIENTE_ENCRIPTADO'].astype('object')
df['COD_SEXO_CALC_BI'] = df['COD_SEXO_CALC_BI'].astype('category')
df['FLAG_SERVICO_BANDA_LARGA'] = df['FLAG_SERVICO_BANDA_LARGA'].astype('bool')
print(df['FLAG_SERVICO_BANDA_LARGA'].head(5))
df['MESES_ANTIGUIDADE'] = df['MESES_ANTIGUIDADE'].astype('int')
df['IDADE'] = df['IDADE'].astype('int')
df['COD_JAVA_EQUIP'] = df['COD_JAVA_EQUIP'].astype('int')
df['COD_ESC2_IDADE_REAL_OU_INF'] = df['COD_ESC2_IDADE_REAL_OU_INF'].astype('int')
df['FLAG_EMAIL'] = df['FLAG_EMAIL'].astype('bool')
df['MM3_VAL_ROAMING'] = df['MM3_VAL_ROAMING'].astype('float')
df['MM3_QTD_CARREG'] = df['MM3_QTD_CARREG'].astype('float')
df['MM3_VAL_CARREG'] = df['MM3_VAL_CARREG'].astype('float')
df['FLAG_INTERNETNOTELEMOVEL'] = df['FLAG_INTERNETNOTELEMOVEL'].astype('bool')
df['FLAG_SERVICO_MEO_MOBILE'] = df['FLAG_SERVICO_MEO_MOBILE'].astype('bool')
df['FLAG_SERVICO_INTERNET_NAO_BL'] = df['FLAG_SERVICO_INTERNET_NAO_BL'].astype('bool')
df['S3_QTD_UTL_INTERNETNOTELEMOVEL'] = df['S3_QTD_UTL_INTERNETNOTELEMOVEL'].astype('int')
df['S3_QTD_KB_INTERNETNOTELEMOVEL'] = df['S3_QTD_KB_INTERNETNOTELEMOVEL'].astype('int')
df['S3_VAL_UTL_INTERNETNOTELEMOVEL'] = df['S3_VAL_UTL_INTERNETNOTELEMOVEL'].astype('float')
df['S3_VAL_UTL_JOGOS_JAVA'] = df['S3_VAL_UTL_JOGOS_JAVA'].astype('float')
```

```python
#para nos certificarmos que estes clientes estão realmente fora
#dos conjuntos  de dados podemos utilizar outra biblioteca que
#é o Z-score
#o retorno indica que estão e qual o valor deste desvio em 
#relação ao conjunto absoluto dos dados
from scipy import stats
z = np.abs(stats.zscore(df['IDADE'].values))
threshold = 3
result = np.where( z > threshold)
df_IDADE_outlier = df.iloc[result[0]]
print(z)
```
